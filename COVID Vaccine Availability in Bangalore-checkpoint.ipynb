```python
### There is a website www.under45.in, one can choose the City and muncipality and you will be directed to a Telegram group, where you
###get notifications everyday about the availability of covid vaccines in hospitals around.
```

```python
### I have joined the group and downloaded the chat history from the telegram group as a JSON file. Loaded the file into pandas and here starts my analysis.
```

```python
import pandas as pd
import json
```

```python
with open ('TelegramChat.json', errors = 'ignore') as data_file:
    data = json.load(data_file)
```

```python
data
```

```python
### Loading the Json file into dataframe
df = pd.DataFrame(data['messages'])
df
```

```python
## Finding null values in columns
df['photo'].isnull().sum()
```

```python
df.columns
```

```python
##Finding null values in columns
df['height'].isnull().sum()
```

```python
### Dropping the columns that contains no data
df.drop(columns = ['photo', 'width', 'height', 'reply_to_message_id',
       'poll'], inplace= True)
df
```

```python
### finding null values in other columns
df['actor'].isnull().sum()
```

```python
df['from_id'].isnull().sum()
```

```python
new_df = df[df['type'] == 'service']
```

```python
new_df
```

```python
## Dropping other columns that contain no data
df.drop(columns = ['actor', 'actor_id', 'action', 'from', 'from_id'], inplace= True)
df
```

```python
df['title'].isnull().sum()
```

```python
## Dropping yet two more columns that contain no data
df.drop(columns = ['type','title'],inplace=True)
df
```

```python
df.info()
```

```python
## Getting to see the data in the column 'text'
display(df.text.iloc[210]) ---->##Here we can see that the 'text' column contains the key information that is the hospital name in which the vaccine is available
```

```python
display(df.text.iloc[163])
```

```python
## Converting text column from object to string format
import warnings
warnings.filterwarnings("ignore")
df['text'] = df['text'].astype(pd.StringDtype(), errors='ignore')
```

```python
## Checking the datatypes
df.info()
```

```python
##In the data, there are messages that exclusively mention the vaccine name. So finding out how many messages exclusively mentioned
##covid vaccine name in India ---->covaxin and covishield
temp_df = df[(df.text.str.contains("COVAXIN")) | (df.text.str.contains("COVISHIELD")) | (df.text.str.contains("COVIDSHIELD"))]
len(temp_df)
```

```python
##Finding out how many message notifications came without the mention of vaccine name.
vaccine_df = df[~((df.text.str.contains("COVAXIN")) | (df.text.str.contains("COVISHIELD")) | (df.text.str.contains("COVIDSHIELD")))]
v = len(vaccine_df)
print(f'There are {v} messages in the group which says vaccine is available without mentioning the vaccine name')
```

```python
## Here we are interested to know EXCLUSIVELY about two particular vaccines. 
## Covaxin and Covishield. So subsetting the dataframe into two EXCLUSIVE groups --- Covaxin and Covishield
covaxin_df = df[df['text'].str.contains("COVAXIN")]
covaxin_df
```

```python
##Checking the datatypes of the Covaxin df
covaxin_df.info()
```

```python
#covaxin_df
writer = pd.ExcelWriter('Covaxin.xlsx')
covaxin_df.to_excel(writer)
writer.save()
```

```python
covidshield_df=df[df['text'].str.contains("COVIDSHIELD")]
covidshield_df
```

```python
##Creating a subset of Covishield
covishield_df = df[df['text'].str.contains("COVISHIELD")]

covishield_df
```

```python
##Checking the datatypes of newly created covishield df
covishield_df.info()
```

```python
a = len(covishield_df)
b = len(covaxin_df)
print(f'Covishield became available {a} times as per the group')
print(f'Covaxin became available {b} times as per the group')
```

```python
##Copying Covishield df to another df to preserve the originial one.
##Splitting the text column to get the hospital info
d2=covishield_df.copy()
d2[['text_1', 'text_2']] = d2["text"].str.split(":", 1, expand=True)
d2
```
