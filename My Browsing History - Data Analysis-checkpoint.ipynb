```python
### I used TLDExtract to get the domain names from the url
import tldextract
tldextract.extract('http://forums.news.cnn.com/')
```

```python
## Creating a new column Domain 
df1['domain'] = None
df1
```

```python
##Writing a function to get the domain.
def get_domain(x):
    domain = tldextract.extract(x)[1]
    sub_domain = tldextract.extract(x)[0]
    if sub_domain == "mail":
        return sub_domain + "." + domain
    # Ugly hack to differentiate b/w drive.google.com and google.com
    if domain == "google" and sub_domain=="www":
        return "google_search" 
    return domain
```

```python
df1['domain'] = df1['url'].apply(get_domain)
df1
```
