```python
import pandas as pd
import re
import regex
import emoji
import matplotlib.pyplot as plt
%matplotlib inline
import datetime
```

```python
###Reading the file: fwf ---> fixed width files
d3=pd.read_fwf("C:\\Users\\User\\Downloads\\Madhavi\\Vangara.txt", colspecs='infer', widths=None, infer_nrows=100,delimiter='\n', header=None, names=['text'])
d3
```

```python
###Splitting the text into two columns: 'datetime_str','text_2' and then dropping the text column
d3[['datetime_str', 'text_2']] = d3["text"].str.split(" - ", 1, expand=True)
d3
```

```python
d3.drop(columns = 'text', inplace=True)
d3
```

```python
###Splitting the 'datetime_str' column into two columns: date, time
d3[['date', 'time']] = d3["datetime_str"].str.split(", ", 1, expand=True)
d3
```

```python
### Since we have split time and date columns separately, we are dropping 'datetime_str' column
d3.drop(columns='datetime_str', inplace=True)
d3
```

```python
d3.dtypes
```

```python
#### converting date column to datetime datatype. The key word is coerce, that will cut the unnecesary spill over text in the date column
d3["date"]= pd.to_datetime(d3["date"], errors='coerce')
d3
```

```python
d3.dtypes
```

```python
##Splitting 'text_2' column into two columns: they are User and Message and then dropping 'text_2' column
d3[['User', 'Message']] = d3["text_2"].str.split(": ", 1, expand=True)
d3
```

```python
d3.drop(columns='text_2', inplace=True)
d3
```

```python
from wordcloud import WordCloud, STOPWORDS
all_words = ' '
stopwords=STOPWORDS.update(['baa', 'ra', 'sare', 'nuvvu', 'yem', 'vaadiki', 'amma','meeru','ippude'])
for msg in d4['Message'].values:
    words=str(msg).lower().split()
    for word in words:
        all_words=all_words + word + ' '
        
wordcloud = WordCloud(width = 1000, height = 800,
                     background_color = 'white',
                     stopwords = stopwords,
                     min_font_size = 10).generate(all_words)

wordcloud.to_image()
```
